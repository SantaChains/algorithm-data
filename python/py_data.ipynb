## Array (数组)

- 随机访问 random_access
- 插入 insert / 删除 delete / 空间分配 space_allocation
- Python: list (动态数组) / array.array (同类型数组)

```python
import array

# list - 动态数组（最常用）
lst = [1, 2, 3, 4, 5]
print(f"访问 lst[2]: {lst[2]}")  # O(1)

lst.insert(2, 99)  # O(n) - 插入
print(f"插入后: {lst}")

lst.pop(2)  # O(n) - 删除指定位置
print(f"删除后: {lst}")

lst.append(6)  # O(1) 均摊 - 追加
print(f"追加后: {lst}")

# array.array - 同类型数组（高效内存）
arr = array.array('i', [1, 2, 3, 4, 5])  # 'i' = signed int
print(f"array.array: {arr}")
```

## Dictionary (字典)

- 键值对 key-value
- 哈希表实现 hash_table
- Python: dict

```python
d = {'a': 1, 'b': 2, 'c': 3}

print(f"访问 d['b']: {d['b']}")  # O(1) 均摊

d['d'] = 4  # O(1) 均摊 - 插入/更新
print(f"插入后: {d}")

print(f"获取 d.get('x', -1): {d.get('x', -1)}")  # O(1) 均摊

del d['c']  # O(1) 均摊 - 删除
print(f"删除后: {d}")

for k, v in d.items():
    print(f"  {k}: {v}")
```

## List (列表)

- 有序可重复序列 ordered_sequence
- Python: list

```python
lst = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"访问 lst[3]: {lst[3]}")  # O(1)

lst.insert(1, 99)  # O(n) - 插入
print(f"插入后: {lst}")

lst.remove(99)  # O(n) - 删除首次出现的值
print(f"remove(99) 后: {lst}")

lst.append(5)  # O(1) 均摊 - 追加
print(f"追加后: {lst}")

lst.extend([7, 8])  # O(k) - 扩展
print(f"extend([7, 8]) 后: {lst}")

print(f"切片 lst[1:4]: {lst[1:4]}")  # O(end-start)

lst.reverse()  # O(n) - 反转
print(f"反转后: {lst}")

lst.sort()  # O(n log n) - 排序
print(f"排序后: {lst}")
```

## Tuple (元组)

- 有序不可变序列 immutable_sequence
- Python: tuple

```python
tpl = (3, 1, 4, 1, 5, 9)

print(f"访问 tpl[2]: {tpl[2]}")  # O(1)

print(f"切片 tpl[1:4]: {tpl[1:4]}")  # O(end-start)

print(f"查找 tpl.index(4): {tpl.index(4)}")  # O(n)
print(f"计数 tpl.count(1): {tpl.count(1)}")  # O(n)

tpl2 = tpl + (2, 7)  # O(n) - 拼接
print(f"拼接后: {tpl2}")

print("注: tuple 不可变，无插入/删除/修改操作")
```

## Linked List (链表)

- 节点寻址 node_addressing
- Python: collections.deque 或自定义 Node 类

```python
from collections import deque

# 使用 deque 模拟链表（双向队列）
dq = deque([1, 2, 3])

dq.appendleft(0)  # O(1) - 头部插入
print(f"头部插入 0: {dq}")

dq.append(4)  # O(1) - 尾部插入
print(f"尾部插入 4: {dq}")

dq.popleft()  # O(1) - 头部删除
print(f"头部删除后: {dq}")

dq.pop()  # O(1) - 尾部删除
print(f"尾部删除后: {dq}")

print("注: 访问需遍历 - O(n)")
```

## Heap (堆)

- 完全二叉树 complete_binary_tree
- 最大堆 max_heap / 最小堆 min_heap
- Python: heapq 模块 (最小堆)

```python
import heapq

heap = [3, 1, 4, 1, 5, 9, 2, 6]
heapq.heapify(heap)  # O(n) - 堆化
print(f"堆化后: {heap}")

heapq.heappush(heap, 0)  # O(log n) - 入堆
print(f"入堆 0 后: {heap}")

min_val = heapq.heappop(heap)  # O(log n) - 出堆
print(f"出堆值: {min_val}, 出堆后: {heap}")

print(f"查看堆顶 heap[0]: {heap[0]}")  # O(1)

print("注: Python 只有最小堆，最大堆需取反存储")
```

## Stack (栈)

- 后进先出 LIFO (Last In First Out)
- Python: list 作为栈

```python
stack = []

stack.append(1)  # 入栈 push - O(1)
stack.append(2)
stack.append(3)
print(f"入栈后: {stack}")

top = stack.pop()  # 出栈 pop - O(1)
print(f"出栈值: {top}, 出栈后: {stack}")

peek = stack[-1]  # 查看栈顶 peek - O(1)
print(f"栈顶: {peek}")

is_empty = not stack  # 判空 is_empty - O(1)
print(f"是否为空: {is_empty}")
```

## Queue (队列)

- 先进先出 FIFO (First In First Out)
- Python: collections.deque (双向队列，高效)

```python
from collections import deque

queue = deque([1, 2, 3])

queue.append(4)  # 入队 enqueue - O(1)
print(f"入队后: {queue}")

front = queue.popleft()  # 出队 dequeue - O(1)
print(f"出队值: {front}, 出队后: {queue}")

print(f"查看队头 queue[0]: {queue[0]}")  # O(1)
print(f"查看队尾 queue[-1]: {queue[-1]}")  # O(1)

is_empty = not queue  # 判空 is_empty - O(1)
print(f"是否为空: {is_empty}")

print("注: list.pop(0) 为 O(n)，不适合做队列")
```
